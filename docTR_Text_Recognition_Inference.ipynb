```python
import os
import torch
import os
from doctr.models import ocr_predictor
from collections import OrderedDict
from doctr.io import DocumentFile
import cv2
import matplotlib.pyplot as plt
```

NOTE: MODEL FILE NOT REQUIRED FOR TEXT RECOGNITION INFERENCE

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_file = '/content/drive/MyDrive/doctr/db_resnet50.pt'

if isinstance(model_file, str):
    predictor = ocr_predictor(pretrained=True)
    state_dict = torch.load(model_file)
    # print(state_dict)
    new_state_dict = OrderedDict()
    for k, v in state_dict.items():
        name = k[7:] # remove `module.`
        new_state_dict[name] = v
        # print(k)
    model = predictor.det_predictor.model
    print(type(model))
    print(model)
    # model.load_state_dict(new_state_dict)
    predictor.det_predictor.model.load_state_dict(new_state_dict)
else:
    predictor = ocr_predictor(pretrained=True)
```

```python
doc = DocumentFile.from_images('/content/MultiCol1OL.png')
result = predictor(doc)
dic = result.export()
```
